```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_grid(m, n, cmap='viridis', diagonal_only=False, gaussian=False):
    # Generate colors from colormap
    colormap = plt.get_cmap(cmap)

    if diagonal_only and m == n:
        colors = np.zeros((m, n))
        np.fill_diagonal(colors, np.random.rand(m))
    elif gaussian:
        i_indices, j_indices = np.indices((m, n))
        shift = np.abs((m - n) // 2)
        if m > n:
            j_indices = j_indices + shift
        elif n > m:
            i_indices = i_indices + shift
        colors = np.exp(-((i_indices - j_indices) ** 2) / 2)
    elif m == n:
        colors = np.random.rand(m, m)
        colors = np.tril(colors) + np.tril(colors, -1).T
    else:
        colors = np.random.rand(m, n)

    fig, ax = plt.subplots(figsize=(n, m))

    for i in range(m):
        for j in range(n):
            color = colormap(colors[i, j])
            rect = plt.Rectangle((j, m - i - 1), 1, 1, facecolor=color, edgecolor='black', linewidth=2)
            ax.add_patch(rect)

    ax.set_xlim(0, n)
    ax.set_ylim(0, m)
    ax.set_aspect('equal')
    ax.axis('off')

    plt.show()
```

```python
# Example usage
plot_grid(5, 8)  # Rectangle grid
plot_grid(6, 6)  # Symmetric square grid
plot_grid(6, 6, diagonal_only=True)  # Diagonal matrix
```

```python
sz1 = 5
sz2 = 8
cmap = 'cividis'
plot_grid(sz1, 1, cmap=cmap)  # Rectangle grid with specified colormap
plot_grid(sz1, sz2, cmap=cmap, gaussian=True)  # Rectangle grid with specified colormap
plot_grid(sz2, sz2, cmap=cmap, gaussian=True)  # Rectangle grid with specified colormap
plot_grid(sz2, sz2, cmap=cmap, diagonal_only=True)  # Rectangle grid with specified colormap
plot_grid(sz2, 1, cmap=cmap)  # Rectangle grid with specified colormap
```
