```python
import numpy as np
import matplotlib.pyplot as plt

# --- True curve and noisy sample ---
np.random.seed(1)
x_true = np.linspace(0, 10, 500)
y_true = np.sin(x_true/2)

# Sparse noisy samples
x_obs = np.linspace(0, 10, 60)
y_obs = np.interp(x_obs, x_true, y_true) + np.random.normal(0, 0.3, size=x_obs.shape)

# --- Define local region (e.g., center at x=5, ±0.8 width) ---
center = 5.0
width = 1
mask_local = np.abs(x_obs - center) < width

# Local mean
y_local_mean = np.mean(y_obs[mask_local])

# --- Plot ---
fig, ax = plt.subplots(figsize=(10, 5))

# True function
ax.plot(x_true, y_true, '--', color='tab:green', label='True signal')

# Full sampled data
ax.plot(x_obs[~mask_local], y_obs[~mask_local], 'o', label='Other samples', color='gray', markersize=8, fillstyle='none')

# Highlight local samples
ax.plot(x_obs[mask_local], y_obs[mask_local], 'o', label='Samples in local mean', color='tab:green', markersize=8)

ax.set_axis_off()
# # Horizontal local mean line
# ax.hlines(y_local_mean, center - width, center + width, colors='tab:green', linestyles='-', linewidth=2, label='Local mean')

# Ver
```
